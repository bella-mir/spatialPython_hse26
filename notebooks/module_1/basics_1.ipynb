## 1. Математические операции

Python можно использовать как калькулятор. Основные операции:


````python
2 + 3    # сложение
````

````python
10 / 4   # деление
````

````python
2 ** 8   # возведение в степень
````

| Операция  | Символ |
| :-------- | :----: |
| Сложение  |  `+`   |
| Вычитание |  `-`   |
| Умножение |  `*`   |
| Деление   |  `/`   |
| Степень   |  `**`  |


## 2. Переменные

**Переменная** — это имя, которому мы присваиваем значение. Удобно давать переменным осмысленные названия в стиле `snake_case`.


````python
temp_celsius = 17.0
temp_fahrenheit = 9 / 5 * temp_celsius + 32

print("Температура в °C:", temp_celsius)
print("Температура в °F:", temp_fahrenheit)
````

Важно помнить: изменение одной переменной **не меняет** другие переменные, которые на неё ссылались раньше.


````python
temp_celsius = 25.0  # меняем значение

print("°C:", temp_celsius)        # → 25.0 (новое значение)
print("°F:", temp_fahrenheit)     # → 62.6 (старое, не пересчиталось!)
````

## 3. Типы данных

У каждого значения в Python есть **тип**. Основные четыре:


| Тип     | Описание            |   Пример   |
| :------ | :------------------ | :--------: |
| `int`   | Целое число         |    `42`    |
| `float` | Дробное число       |   `3.14`   |
| `str`   | Строка (текст)      | `'Москва'` |
| `bool`  | Логическое значение |   `True`   |


````python
city = "Москва"
population = 13_104_177
area_km2 = 2561.5
is_capital = True

print(type(city))        # <class 'str'>
print(type(population))  # <class 'int'>
print(type(area_km2))    # <class 'float'>
print(type(is_capital))  # <class 'bool'>
````

### Преобразование типов

Если нужно смешать данные разных типов, их сначала нужно привести к одному типу.


````python
value_str = "3.14"       # строка, выглядящая как число
value_float = float(value_str)  # преобразуем в float

print(value_float + 1)   # → 4.14
````

````python
# int() отбрасывает дробную часть (не округляет!)
print(int(3.99))   # → 3

# round() — выполняет математическое округление
print(round(3.99)) # → 4
````

## 4. Списки

**Список** (`list`) позволяет хранить несколько значений под одним именем. Элементы разделяются запятыми и заключаются в квадратные скобки.


````python
cities = ["Москва", "Санкт-Петербург", "Казань", "Новосибирск"]

print(cities)        # весь список
print(len(cities))   # длина → 4
````

### Индексация

Доступ к элементам осуществляется по **индексу** — порядковому номеру, начиная с `0`. Отрицательные индексы считают с конца.


````python
print(cities[0])   # Москва (первый)
print(cities[2])   # Казань (третий)
print(cities[-1])  # Новосибирск (последний)
````

### Изменение списка


````python
# Добавить элемент в конец
cities.append("Екатеринбург")

# Изменить элемент
cities[1] = "СПб"

# Удалить элемент
del cities[1]

print(cities)
````

### Полезные методы списков

| Метод        | Что делает                                  |
| :----------- | :------------------------------------------ |
| `.append(x)` | Добавляет `x` в конец списка                |
| `.remove(x)` | Удаляет первое вхождение `x`                |
| `.pop(i)`    | Удаляет и возвращает элемент с индексом `i` |
| `.sort()`    | Сортирует список по возрастанию             |
| `.reverse()` | Разворачивает список                        |
| `.index(x)`  | Возвращает индекс элемента `x`              |
| `.count(x)`  | Считает количество вхождений `x`            |


````python
nums = [5, 2, 8, 1, 9]
nums.sort()
print(nums)   # [1, 2, 5, 8, 9]
````

> ⚠️ Методы `.sort()` и `.reverse()` изменяют список **на месте** и возвращают `None`. Не пишите `my_list = my_list.sort()` — так вы потеряете данные!


## 5. Форматирование строк

Самый современный и удобный способ — **f-строки** (f-strings). Перед кавычками ставится буква `f`, а переменные подставляются прямо в текст через `{}`.


````python
station = "Казань"
lat = 55.7963
lon = 49.1088
temp = 21.456789

# Вставляем переменные, температуру округляем до 1 знака
info = f"Станция {station} ({lat:.4f}°N, {lon:.4f}°E): температура {temp:.1f}°C"
print(info)
````

Форматирование чисел внутри f-строк:

| Формат | Значение              | Пример     |
| :----- | :-------------------- | :--------- |
| `:.2f` | 2 знака после запятой | `3.14`     |
| `:.4f` | 4 знака после запятой | `55.7963`  |
| `:.2e` | Научная нотация       | `3.14e+00` |
| `:.1%` | Проценты              | `31.4%`    |


### Основные операции со строками


````python
text = "  Привет, мир!  "

print(text.strip())           # убрать пробелы по краям → "Привет, мир!"
print(text.upper())           # верхний регистр
print(text.lower())           # нижний регистр
print(text.replace("мир", "Python"))  # замена
print(text.split(","))        # разбить по символу → список
````

````python
# Срезы строк работают как индексы у списков
name = "Новосибирск"
print(name[0])    # Н
print(name[0:4])  # Ново
print(name[-2:])  # ск
````

## 6. Цикл for

Цикл `for` позволяет выполнять одни и те же действия для каждого элемента коллекции — например, списка. Это избавляет от необходимости писать одинаковый код многократно.


Без цикла:


````python
cities = ["Москва", "Казань", "Новосибирск"]

print(cities[0])
print(cities[1])
print(cities[2])
````

С циклом:


````python
for city in cities:
    print(city)
````

### Синтаксис цикла

```python
for переменная in коллекция:
    # код цикла
```

Важные правила:

- после `:` нужен отступ;
- обычно используется 4 пробела;
- код с одинаковым отступом относится к одному блоку.


````python
coords = [55.75, 51.82, 56.85, 43.99]

for lat in coords:
    print(f"Широта: {lat:.2f}°")
````

### Функция `range()`

`range()` создаёт последовательность чисел.


````python
for i in range(5):
    print(i)
````

````python
for i in range(0, 100, 25):
    print(i)
````

### Цикл по индексам

Иногда нужен доступ к индексам элементов.


````python
cities = ["Москва", "Санкт-Петербург", "Казань"]
populations = [13_104_177, 5_601_000, 1_308_660]

for i in range(len(cities)):
    print(f"{cities[i]}: {populations[i]:,} чел.")
````

### Изменение элементов списка


````python
lats = [55.7522, 59.9386, 56.8519]

for i in range(len(lats)):
    lats[i] = round(lats[i], 1)

print(lats)
````

## 7. Условные операторы

Условные операторы позволяют выполнять разный код в зависимости от условия.

### Конструкция `if / else`

Базовая конструкция: если условие истинно — выполняется блок `if`, иначе — блок `else`.


````python
temperature = 17

if temperature > 25:
    print(f"{temperature}°C — жарко!")
else:
    print(f"{temperature}°C — не жарко.")
````

Блок `else` необязателен — если условие ложно и `else` нет, Python просто ничего не делает.

Как и в циклах, тело условного блока выделяется **отступом 4 пробела**, после условия ставится **двоеточие `:`**.


### Операторы сравнения

| Оператор | Описание         |
| :------: | :--------------- |
|   `==`   | Равно            |
|   `!=`   | Не равно         |
|   `<`    | Меньше           |
|   `>`    | Больше           |
|   `<=`   | Меньше или равно |
|   `>=`   | Больше или равно |

Результат сравнения — это **булево значение** (`True` или `False`).


````python
print(10 > 5)
print(10 == 5)
print("GIS" != "gis")
````

### `if / elif / else`

Для проверки нескольких условий подряд используется `elif` (сокращение от «else if»). Python проверяет условия по порядку и выполняет первый подходящий блок.


````python
temperature = -3

if temperature > 0:
    print("Выше нуля")
elif temperature == 0:
    print("Точка замерзания")
else:
    print("Ниже нуля")
````

### Логические операторы

Несколько условий можно объединять с помощью `and`, `or`, `not`:

| Оператор | Результат `True`, если...    |
| :------: | :--------------------------- |
|  `and`   | Оба условия истинны          |
|   `or`   | Хотя бы одно условие истинно |
|  `not`   | Условие ложно                |


````python
weather = "дождь"
wind_speed = 14

if weather == "дождь" or wind_speed >= 10:
    print("Лучше остаться дома")
````

### Условие внутри цикла

Условные операторы часто используются внутри циклов — это один из самых распространённых паттернов в анализе данных.


````python
temperatures = [-5, 0, 12, 25, 31]

for temp in temperatures:
    if temp > 25:
        label = "жарко"
    elif temp > 10:
        label = "тепло"
    elif temp > 0:
        label = "прохладно"
    else:
        label = "мороз"

    print(f"{temp:>4}°C — {label}")
````
