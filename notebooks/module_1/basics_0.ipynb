# I. Основы Python

## Темы

1. [**Введение в Python**](basics_1.ipynb)

   Познакомитесь с базовыми возможностями Python: переменными, типами данных, математическими операциями и списками.

2. [**Функции, модули и библиотеки**](basics_2.ipynb)

   Научитесь создавать собственные функции, работать с параметрами и возвращаемыми значениями, использовать модули и документировать код.

3. [**Основы pandas**](basics_3.ipynb)

   Познакомитесь с библиотекой `pandas`, научитесь читать табличные данные, анализировать их и выполнять базовую обработку.
