# Функции, модули и библиотеки

В этом разделе мы научимся оформлять повторяющийся код в функции, документировать его, а также подключать готовые модули и библиотеки Python.


## 1. Функции

**Функция** — это именованный блок кода, который выполняет конкретную задачу и может использоваться многократно. Хорошее правило программирования: не повторяй себя — если один и тот же код нужен несколько раз, оформи его как функцию.


### Создание функции

Функция определяется ключевым словом `def`, за которым следуют имя, параметры в скобках и двоеточие. Тело функции выделяется отступом. Оператор `return` задаёт возвращаемое значение.


```python
def celsius_to_kelvin(temp_celsius):
    return temp_celsius + 273.15
```

```python
# Вызываем функцию
print(celsius_to_kelvin(0))
print(celsius_to_kelvin(100))
```

### Вызов функции внутри функции

Функции можно строить из других функций — это делает код понятным и легко поддерживаемым.


```python
def celsius_to_fahr(temp_celsius):
    return 9 / 5 * temp_celsius + 32

def kelvin_to_fahr(temp_kelvin):
    temp_celsius = temp_kelvin - 273.15
    return celsius_to_fahr(temp_celsius)  # используем уже готовую функцию

print(f"0 K = {kelvin_to_fahr(0):.2f}°F")
```

### Параметры функции

Параметрам можно задавать **значения по умолчанию** — тогда они становятся необязательными. Обязательные параметры всегда указываются **перед** необязательными.


```python
def describe_point(lat, lon, name="точка"):
    """
    Выводит описание географической точки.

    Parameters
    ----------
    lat : float
        Широта в градусах.
    lon : float
        Долгота в градусах.
    name : str, optional
        Название точки (по умолчанию "точка").
    """
    print(f"{name}: {lat:.4f}°N, {lon:.4f}°E")

describe_point(55.7522, 37.6156, "Москва")  # с именем
describe_point(59.9386, 30.3141)            # без имени — используется значение по умолчанию
```

При вызове функции можно явно указывать имена параметров — тогда их порядок не важен:


```python
describe_point(name="Казань", lon=49.1088, lat=55.7963)
```

### Несколько возвращаемых значений

Функция может возвращать несколько значений через запятую — Python упакует их в кортеж, который можно сразу распаковать в переменные.


```python
def bbox_center(min_lat, max_lat, min_lon, max_lon):
    """Возвращает координаты центра ограничивающего прямоугольника."""
    center_lat = (min_lat + max_lat) / 2
    center_lon = (min_lon + max_lon) / 2
    return center_lat, center_lon

# Распаковываем результат сразу в две переменные
lat, lon = bbox_center(55.1, 56.3, 36.8, 38.2)
print(f"Центр: {lat:.2f}°N, {lon:.2f}°E")
```

### Область видимости переменных

Переменные, определённые **внутри** функции, существуют только в ней — они недоступны снаружи. Это называется локальной областью видимости.


```python
def compute_area(width, height):
    area = width * height  # локальная переменная
    return area

result = compute_area(3, 4)
print(result)  # 12

# print(area)  # ← NameError: area не существует вне функции
```

### Документирование функций (docstring)

Хорошей практикой является добавление **docstring** — строки документации сразу после `def`. Она объясняет, что делает функция, какие принимает параметры и что возвращает.


```python
def haversine_approx(lat1, lon1, lat2, lon2):
    """
    Приближённое расстояние между двумя точками на сфере (в км).

    Parameters
    ----------
    lat1, lon1 : float
        Координаты первой точки в градусах.
    lat2, lon2 : float
        Координаты второй точки в градусах.

    Returns
    -------
    float
        Расстояние в километрах.
    """
    import math
    R = 6371  # радиус Земли в км
    dlat = math.radians(lat2 - lat1)
    dlon = math.radians(lon2 - lon1)
    a = math.sin(dlat/2)**2 + math.cos(math.radians(lat1)) * math.cos(math.radians(lat2)) * math.sin(dlon/2)**2
    return R * 2 * math.asin(math.sqrt(a))

# Расстояние Москва → Казань
dist = haversine_approx(55.7522, 37.6156, 55.7963, 49.1088)
print(f"Москва → Казань: {dist:.0f} км")
```

```python
# Просмотр документации функции
help(haversine_approx)
```

## 2. Модули и библиотеки

**Модуль** — это файл с готовым кодом (функциями, константами), который можно подключить к своей программе. Несколько модулей объединяются в **пакет**, а пакеты — в **библиотеку**. На практике эти слова часто используются как синонимы.

Именно богатая экосистема библиотек делает Python таким популярным инструментом для анализа данных и ГИС.


### Способы импорта

#### 1. Стандартный импорт


```python
import math
```

```python
print(math.sqrt(81))   # → 9.0
print(math.pi)         # → 3.14159...
```

#### 2. Импорт с псевдонимом (alias) — удобно для длинных имён


```python
import numpy as np          # стандартный псевдоним для NumPy
import pandas as pd         # стандартный псевдоним для pandas
import geopandas as gpd     # стандартный псевдоним для GeoPandas
```

```python
print(type(np)) 
```

#### 3. Импорт конкретной функции из модуля


```python
from math import sqrt, radians

print(sqrt(144))         # → 12.0
print(radians(180))      # → 3.14159...
```

```python
# 4. Импорт подмодуля
import matplotlib.pyplot as plt

plt.plot([1, 2, 3], [3, 1, 2])
plt.title("Пример графика")
plt.show()
```

> ⚠️ Не используйте `from module import *` — это засоряет пространство имён и затрудняет понимание кода.


### Встроенные функции
