# VII. Интерактивная визуализация


## Темы

1. [**Интерактивная карта**](map_1.ipynb)

   Научитесь создавать веб-карты с помощью Folium: добавлять несколько слоёв (землепользование, остановки, зоны доступности), настраивать стили, всплывающие подсказки и элементы управления.

2. [**Форматы визуализации**](map_2.ipynb)

   Познакомитесь с разными способами визуализации данных в Folium: подложки, маркеры и иконки, всплывающие окна, легенды, окраска по значению (colormap и Choropleth), кластеры и тепловые карты, а также плагины DualMap, Draw и AntPath.

3. [**Публикация на gh-pages**](map_3.ipynb)

   Узнаете, как сохранить карту в HTML и разместить её в интернете с помощью GitHub Pages, чтобы делиться результатами анализа по ссылке.
