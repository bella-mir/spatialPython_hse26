# II. Пространственные данные


## Темы


1. [**Модели пространственных данных**](spData_1.ipynb)

   Познакомитесь с двумя основными типами представления пространственных данных — векторным и растровым. Узнаете, в каком виде хранится информация и какие классы в Python используются для их описания.

2. [**Чтение векторных данных**](spData_2.ipynb)

   Научитесь читать векторные данные из различных форматов и создавать набор пространственных данных на основе координат и сохранять результат.

3. [**Изучение набора данных**](spData_3.ipynb)

   Узнаете, как получить информацию о системе координат, количестве объектов, типе геометрии, полях и типах данных в них и других характеристиках.

4. [**Экспорт данных из OSM**](spData_4.ipynb)

   Научитесь экспортировать данные из OpenStreetMap по тегам и их значениям с помощью библиотеки `osmnx`.
