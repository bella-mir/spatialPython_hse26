Есть вариант выгрузить данные внутри полигона через `ox.features_from_polygon()` (например, если выше данные не выгружались из-за их объема)

Ниже дана функция, которая создаст полигон 1км на 1км на основе заданных вами координат в градусах. Например, можете задать координаты центра города, который хотите изучить. Чтобы воспользоваться функций, вам нужно просто раскомментировать код.


```python
from shapely.geometry import box
import geopandas as gpd

def bbox_from_center(lat, lon, size_m=1500):
    """
    Создаёт квадратный полигон size_m × size_m вокруг точки (lat, lon).
    Координаты возвращаются в EPSG:4326.
    """
    point = gpd.GeoDataFrame(
        geometry=gpd.points_from_xy([lon], [lat]),
        crs="EPSG:4326"
    )
    utm_crs = point.estimate_utm_crs()
    pt_utm = point.to_crs(utm_crs)
    cx, cy = pt_utm.geometry.x.iloc[0], pt_utm.geometry.y.iloc[0]
    half = size_m / 2
    square = box(cx - half, cy - half, cx + half, cy + half)
    return gpd.GeoDataFrame(geometry=[square], crs=utm_crs).to_crs("EPSG:4326")

```
