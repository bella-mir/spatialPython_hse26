# III. Основные инструменты геообработки


## Темы


1. [**Базовые операции с геометрией**](geoprocessing_1.ipynb)

   Познакомитесь с основными инструментами геообработки для работы с геометрией.

2. [**Пространственные отношения**](geoprocessing_2.ipynb)

   Научитесь определять пространственные отношения между объектами, например `intersects`, `contains` или `within`. Освоите фильтрацию данных по пространственному признаку.

3. [**Пространственное объединение**](geoprocessing_3.ipynb)

   Узнаете, как объединять пространственные данные с помощью `sjoin` на основе пространственных отношений.

4. [**Агрегирование точечных данных по регулярной сетке**](geoprocessing_4.ipynb)

   Построите регулярную сетку и подсчитаете количество объектов в ячейках.

5. [**Агрегирование атрибутивной информации**](geoprocessing_5.ipynb)

   Научитесь рассчитывать статистику для объединённых данных на основе атрибутивной информации.
