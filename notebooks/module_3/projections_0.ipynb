# II. Проекции и системы координат


## Темы


1. [**Введение в проекции и системы координат**](projections_1.ipynb)

   Познакомитесь с основными типами систем координат — географическими и проецированными. Узнаете, что такое система координат (CRS), зачем она нужна и как посмотреть информацию о ней в пространственных данных.

2. [**Перепроецирование данных**](projections_2.ipynb)

   Научитесь изменять систему координат пространственных данных с помощью GeoPandas и определять подходящую систему координат для анализа.

3. [**Измерение площадей и расстояний**](projections_3.ipynb)

   Узнаете, как система координат влияет на пространственные измерения и научитесь рассчитывать расстояния и площади в метрах, используя проецированные данные.


## Задание: Изучаем землепользование

Каждый модуль заканчивается практическим заданием — этапом проекта **«15-минутный город»**, посвящённого анализу доступности повседневных функций. Проект можно выполнять для любого интересующего вас города.

В этом модуле мы изучаем структуру землепользования выбранного района. Это поможет понять, как распределены жилые, коммерческие, общественные и рекреационные территории и как структура городской среды влияет на доступность повседневных функций.

Задание оформлено в виде **Jupyter Notebook** с примерной последовательностью шагов выполнения и доступно на платформе [Boosty](https://boosty.to/bellamir).

Там же размещены видео к каждому блоку и есть возможность задавать вопросы. Поддерживая проект, вы помогаете выпуску открытых материалов.
