# IV. Статическая визуализация


## Темы


1. [**Базовые карты и графики**](visualizations_1.ipynb)

   Научитесь визуализировать пространственные данные с помощью метода `.plot()` в GeoPandas и строить статистические графики с помощью `matplotlib`. Познакомитесь с компоновкой: как разместить карту и диаграмму на одном листе и синхронизировать цвета между ними. В конце — знакомство с картографическими подложками через `contextily`.

2. [**Многослойные карты и подложки**](visualizations_2.ipynb)

   Освоите наложение нескольких слоёв данных на одну карту: полигонов, линий и точек. Научитесь строить единую легенду для нескольких слоёв и добавлять картографическую подложку с помощью `contextily`.
