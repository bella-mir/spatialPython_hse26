# Дополнительные материалы

Здесь собраны дополнительные темы, которые рассматриваются за пределами основной программы курса. Раздел будет постепенно пополняться новыми материалами.


## Темы

1. [**Настройка OSMnx**](add_1.ipynb)

   Как настраивать (или попытаться настроить) библиотеку для стабильной выгрузки данных.

2. [**Геокодирование**](add_2.ipynb)

   Как превращать адреса в координаты через API (на примере Yandex Геокодера) и геокодировать целые таблицы.

3. [**Экспорт данных из DATA_MOS**](add_3.ipynb)

   Как забирать открытые данные Москвы через API портала data.mos.ru и собирать из них набор пространственных данных.

4. [**Картографирование статистических данных**](add_4.ipynb)

   Методы классификации данных для карт и картографирование показателей БДМО Росстата по муниципалитетам.
